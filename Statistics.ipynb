```python
import numpy as np

# Define the maximum number of dice rolls
max_rolls = 1000000

# Generate a large array of random integers between 1 and 6
random_integers = np.random.randint(1, 7, size=max_rolls)

# Compute the cumulative sum of the random integers
cumulative_sums = np.cumsum(random_integers)

# Calculate the mean for each subset
mean_values = cumulative_sums / np.arange(1, max_rolls + 1)
```

```python
idx = np.linspace(1, 1000000, 1000000)
```

```python
import matplotlib.pyplot as plt
```

```python
x_coord = len(mean_values) - 1
y_coord = mean_values[-1]

plt.figure(figsize=(14, 4))
plt.plot(idx, mean_values, color="lime")
plt.xlabel("Number of Trials")
plt.ylabel("Outcomes on a Dice")
plt.title("How E(x) is tending toward 3.5 as trial increases")
plt.style.use("dark_background")
plt.annotate(
    text="Expected Value 3.5",
    xy=(x_coord, y_coord),
    xytext=(x_coord + 500, y_coord - 0.1),  # Adjust these values as needed for clarity
    xycoords="data",
    arrowprops=dict(facecolor="black", arrowstyle="->"),
    fontsize=10,
    color="cyan",
)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
```

```python
# Total number of dice rolls
num_rolls = 10000

# Generate random dice rolls (values between 1 and 6)
rolls = np.random.randint(1, 7, size=num_rolls)

# Compute cumulative sums and cumulative means
cumulative_sums = np.cumsum(rolls)
cumulative_means = cumulative_sums / np.arange(1, num_rolls + 1)
```

```python
# Create a figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Set axis labels and title
ax.set_xlabel("Number of Rolls")
ax.set_ylabel("Cumulative Mean")
ax.set_title("Convergence of Cumulative Mean to Expected Value (3.5)")

# Set axis limits
ax.set_xlim(0, num_rolls)
ax.set_ylim(1, 6)

# Plot the expected value line
ax.axhline(y=3.5, color="red", linestyle="--", label="Expected Value (3.5)")

# Initialize the line that will be updated during the animation
(line,) = ax.plot([], [], label="Cumulative Mean")

# Add a legend
ax.legend()
```

```python
# Initialization function: sets the background of each frame
def init():
    line.set_data([], [])
    return (line,)


# Update function: updates the data for each frame
def update(frame):
    x_data = np.arange(1, frame + 1)
    y_data = cumulative_means[:frame]
    line.set_data(x_data, y_data)
    return (line,)
```

```python
# Create the animation
ani = FuncAnimation(
    fig, update, frames=num_rolls, init_func=init, blit=True, interval=1
)

# Display the animation
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Total number of dice rolls
num_rolls = 10000

# Generate random dice rolls (values between 1 and 6)
rolls = np.random.randint(1, 7, size=num_rolls)

# Compute cumulative sums and cumulative means
cumulative_sums = np.cumsum(rolls)
cumulative_means = cumulative_sums / np.arange(1, num_rolls + 1)

# Set up the figure and axis
fig, ax = plt.subplots(figsize=(10, 6))
ax.set_xlim(1, num_rolls)
ax.set_ylim(1, 6)
ax.set_xlabel("Number of Rolls")
ax.set_ylabel("Cumulative Mean")
ax.set_title("Convergence of Cumulative Mean to Expected Value (3.5)")
ax.axhline(y=3.5, color="red", linestyle="--", label="Expected Value (3.5)")
(line,) = ax.plot([], [], label="Cumulative Mean")
ax.legend()


# Initialization function: plot the background of each frame
def init():
    line.set_data([], [])
    return (line,)


# Animation function: update the plot
def update(frame):
    x_data = np.arange(1, frame + 1)
    y_data = cumulative_means[:frame]
    line.set_data(x_data, y_data)
    return (line,)


# Create the animation
ani = FuncAnimation(
    fig, update, frames=num_rolls, init_func=init, blit=True, interval=1
)

# Save the animation as an MP4 file
# ani.save("dice_roll_convergence.mp4", writer="ffmpeg", fps=30)

# To display the animation inline (e.g., in a Jupyter Notebook), use the following:
from IPython.display import HTML

HTML(ani.to_jshtml())

# Show the plot
plt.show()
```
