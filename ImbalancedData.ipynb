```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_classification
```

```python
X, y = make_classification(
    n_samples=10000,
    n_features=20,
    n_informative=2,
    n_redundant=10,
    n_clusters_per_class=1,
    weights=[0.9, 0.1],  # 90% of class 0, 10% of class 1
    flip_y=0,
    random_state=42,
)

# Convert to DataFrame
new_df = pd.DataFrame(X, columns=[f"feature_{i}" for i in range(X.shape[1])])
new_df["target"] = y
```

```python
new_df["target"].value_counts()
```

```python
from imblearn.over_sampling import SMOTE
```

```python
Sampler = SMOTE()
```

```python
new_x, new_y = Sampler.fit_resample(X, y)
```

```python
new_new_df = pd.DataFrame(new_x, columns=[f"feature_{i}" for i in range(new_x.shape[1])])
new_new_df["target"] = new_y
```

```python
new_new_df["target"].value_counts()
```

```python
import matplotlib.pyplot as plt
```

```python
# Scatter plot: feature_0 vs feature_1 colored by target class
plt.figure(figsize=(8, 6))
plt.scatter(
    new_df[new_df["target"] == 0]["feature_0"],
    new_df[new_df["target"] == 0]["feature_1"],
    label="Class 0",
    alpha=0.7,
    s=10,
)
plt.scatter(
    new_df[new_df["target"] == 1]["feature_0"],
    new_df[new_df["target"] == 1]["feature_1"],
    label="Class 1",
    alpha=0.8,
    s=10,
    color="red",
)
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
plt.title("Scatter Plot of Feature 0 vs Feature 1 by Class")
plt.legend()
plt.grid(True)
plt.show()
```
